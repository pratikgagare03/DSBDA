```python
# import module
import requests
from bs4 import BeautifulSoup
```

```python
HEADERS={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}
```

```python
def getdata(site_url):
    res = requests.get(site_url, headers=HEADERS)
    return res.text
```

```python
def gethtml(site_url):
 # pass the url
 # into getdata function
    data = getdata(site_url)
    soup = BeautifulSoup(data, 'html.parser')
 # display html code
    return (soup)
```

```python
site_url = "https://www.amazon.in/Nothing-Phone-White-256-RAM/dp/B0BKZSVF27/ref=sr_1_1?keywords=nothing%2Bphone%2B1&qid=1683702350&sprefix=noth%2Caps%2C294&sr=8-1&th=1"
soup = gethtml(site_url)
print(soup)
```

```python
customer_list = []
for item in soup.find_all("span", class_="a-profile-name"):
    name = item.get_text()
    customer_list.append(name)
```
